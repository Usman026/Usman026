```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from collections import defaultdict
dataset = pd.read_csv('/content/nike_data_2022_09.csv')
column_to_clean = 'description'
dataset[column_to_clean].fillna('', inplace=True)
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
inverted_index = defaultdict(set)

def preprocess_text(text):
    # Lowercasing
    text = text.lower()

    # Removing punctuation and tokenization
    tokens = word_tokenize(re.sub(r'[^\w\s]', ' ', text))

    # Stopword removal and stemming
    tokens = [stemmer.stem(token) for token in tokens if token not in stop_words]

    return tokens
for index, row in dataset.iterrows():
    tokens = preprocess_text(row[column_to_clean])
    for token in tokens:
        inverted_index[token].add(index)
for term, doc_ids in inverted_index.items():
    print(f"Term: {term}, Documents: {doc_ids}")
term_to_doc_map = {term: list(doc_ids) for term, doc_ids in inverted_index.items()}
for term, doc_ids in term_to_doc_map.items():
    print(f"Term: {term}, Documents: {doc_ids}")
```

```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
dataset = pd.read_csv('/content/nike_data_2022_09.csv')
nltk.download('punkt')
stemmer = PorterStemmer()
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', str(text)).lower()
    tokens = word_tokenize(text)
    tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(tokens)
dataset['processed_description'] = dataset['description'].apply(preprocess_text)
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(dataset['processed_description'])
while True:
    query = input("Enter your search query: ")
    processed_query = preprocess_text(query)
    query_vector = tfidf_vectorizer.transform([processed_query])
    similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()
    top_matches = similarities.argsort()[::-1]
    print("\nSearch results:")
    for idx in top_matches[:5]:
        print(f"- {dataset['name'][idx]} (Similarity: {similarities[idx]:.4f})")
    another_query = input("\nDo you want to perform another search? (yes/no): ")
    if another_query.lower() != 'yes':
        print("Goodbye!")
        break
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import precision_score, recall_score, f1_score
dataset = pd.read_csv('/content/nike_data_2022_09.csv')
column_for_retrieval = 'description'
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', str(text)).lower()
    return text
dataset[column_for_retrieval] = dataset[column_for_retrieval].apply(preprocess_text)
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(dataset[column_for_retrieval])
test_queries = ["shoes reviews", "t-shirt", "shocks"]
relevant_documents = [[2, 5, 8], [1, 6, 12], [3, 9, 15]]
for i, query in enumerate(test_queries):
    user_query = preprocess_text(query)
    user_query_vector = tfidf_vectorizer.transform([user_query])

    similarities = cosine_similarity(user_query_vector, tfidf_matrix).flatten()
    ranked_documents = sorted([(idx, sim) for idx, sim in enumerate(similarities)], key=lambda x: x[1], reverse=True)

    retrieved_indices = [idx for idx, _ in ranked_documents[:len(relevant_documents[i])]]
    relevant_indices = relevant_documents[i]
    precision = precision_score(relevant_indices, retrieved_indices, average='micro')
    recall = recall_score(relevant_indices, retrieved_indices, average='micro')
    f1 = f1_score(relevant_indices, retrieved_indices, average='micro')

    print(f"Query: {query}")
    print(f"Retrieved documents: {retrieved_indices}")
    print(f"Relevant documents: {relevant_indices}")
    print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")
    print()
```



### Incorporating User Feedback Mechanisms in Information Retrieval:

**What are User Feedback Mechanisms?**
User feedback mechanisms involve creating avenues for users to provide input, opinions, and evaluations about the relevance and quality of the retrieved documents in an information retrieval system.

**Importance of User Feedback:**
User feedback is crucial because it bridges the gap between the system's algorithms and the actual needs and preferences of users. It allows the system to continuously adapt and improve its performance based on real user interactions.

**How User Feedback Improves Performance:**
1. **Enhancing Relevance**: Users' evaluations provide insights into which documents are relevant to their queries and needs. This information helps the system better rank and retrieve documents in the future.

2. **Personalization**: Feedback helps the system understand individual user preferences, enabling it to tailor results to each user's unique requirements.

3. **Adaptation to Trends**: Feedback patterns can reveal emerging trends and changes in user preferences, enabling the system to stay up-to-date and relevant.

**Types of User Feedback Mechanisms:**
1. **Explicit Feedback**: Users explicitly indicate the relevance of retrieved documents. For example, users can mark documents as "useful" or "not useful."

2. **Implicit Feedback**: User behavior, such as clicks, dwell time, and scroll patterns, can be analyzed to infer document relevance.

3. **Ratings and Reviews**: Users can provide ratings and written reviews for documents, helping both the system and other users assess document quality.

**Benefits of Incorporating User Feedback:**
- **Continuous Improvement**: Feedback-driven adjustments lead to a system that evolves and becomes more aligned with user expectations over time.
- **User-Centricity**: User feedback keeps the system centered on user needs, enhancing user satisfaction.
- **Accurate Relevance**: As user preferences are learned, the system can rank documents more accurately based on what users find relevant.

**Considerations and Challenges:**
- **Data Privacy**: Handle user feedback data responsibly, ensuring privacy and security.
- **Feedback Collection**: Design a user-friendly and non-intrusive mechanism for collecting feedback.
- **Quality Control**: Implement measures to filter out spam or unhelpful feedback that could skew results.

**Code Implementation:**
The example code previously provided demonstrates a basic way to incorporate user feedback. After presenting retrieved documents, the system asks users to mark relevant documents. The feedback influences the calculation of metrics like Precision, Recall, and F1-score, which guide system improvements.

**Long-Term Impact:**
Incorporating user feedback mechanisms results in an information retrieval system that learns, adapts, and becomes more valuable to users over time. It's a way to ensure that the system remains dynamic and user-centric in an ever-changing digital landscape.

By considering user feedback as an integral part of your system's design, you're not only building a more effective tool but also creating a platform that truly serves its users' information needs.
